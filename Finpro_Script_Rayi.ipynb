## Common Financial Analysis

### Returns

```python
#Assigning 'adj close' to 'daily close'
daily_close = aapl[['Adj Close']]

# calculating the daily_returns based on the percentage change in adjusted close
daily_pct_c = daily_close.pct_change()

#replacing the na values with 0
daily_pct_c.fillna(0, inplace = True)

# inspecting daily returns
print(daily_pct_c)

#importing numpy package
import numpy as np

# calculating daily log returns
daily_log_returns = np.log(daily_close.pct_change() + 1)

#print daily log returns
print(daily_log_returns)
```

```python
#resample 'aapl' to 'business months', take last observation as value
monthly = aapl.resample('BM').apply(lambda x: x[-1])

#calculating the monthly percentage change
monthly.pct_change()

# resample 'aapl' to quarters, take the mean as value per quarter
quarter = aapl.resample("4M").mean()

# calculate the quarterly percentage change
quarter.pct_change()
```

```python
#daily returns
daily_pct_c = daily_close / daily_close.shift(1) - 1

# printing our daily returns
print(daily_pct_c)
```

```python
# importing matplotlib package to do some visualizing in our project
import matplotlib.pyplot as plt

# plotting the distribution of daily_pct_c
daily_pct_c.hist(bins=50)

#show the plot
plt.show()

#pulling up summary statistics
print(daily_pct_c.describe())
```

```python
# calculating the cumulative daily returns
cum_daily_returns = (1 + daily_pct_c).cumprod()

# printing the cumulative daily returns
print(cum_daily_returns)
```

```python
def get(tickers, startdate, enddate):
    """Concatenates stock data and produces a data frame.
    
       It accepts ticker or stock symbol along with the start and end dates of stocks as parameters."""
    def data(ticker):
        """Returns the retrieved stock data based on tocker, start date and end date as specified by the user.
        
           Accepts tickers or stock symbol as parameter."""
        return(pdr.get_data_yahoo(ticker, start = startdate, end = enddate))
    datas = map(data, tickers)
    return (pd.concat(datas, keys=tickers, names = ['Ticker', 'Date']))

#storing the stock tickers we want to retrieve stock data about in a list
tickers = ['AAPL', 'MSFT', 'IBM', 'GOOG']

#creating a variable and calling the function based on our tickers, start date and end date
all_data = get(tickers, datetime.datetime(2014, 4, 1), datetime.datetime(2019, 4, 1))

#printing the first few rows of our data frame that has been generated
all_data.head()
```

```python
# we are storing the adjusted close of all tickers in a variable and resetting our index
daily_close_px = all_data[['Adj Close']].reset_index().pivot('Date', 'Ticker', 'Adj Close')

# printing our new dataframe with adjusted close values for all tickers
print(daily_close_px)

#calculating the daily percentage change for 'daily_close_px'
daily_pct_change = daily_close_px.pct_change()

#plot the distributions
daily_pct_change.hist(bins=50, sharex=True, figsize=(12, 8))

#show the resulting plot
plt.show()
```

### Ordinary Least Squares Regression

```python
# import the 'api' model of 'statsmodel' under alias 'sm'
import statsmodels.api as sm
from pandas import tseries
from pandas.core import datetools

# isolate the adjusted closing price
all_adj_close = all_data[['Adj Close']]

# calculate the returns
all_returns = np.log(all_adj_close / all_adj_close.shift(1))

# isolate the AAPL returns
aapl_returns = all_returns.iloc[all_returns.index.get_level_values('Ticker') == 'AAPL']
aapl_returns.index = aapl_returns.index.droplevel('Ticker')

# isolate the MSFT returns
msft_returns = all_returns.iloc[all_returns.index.get_level_values('Ticker') == 'MSFT']
msft_returns.index = msft_returns.index.droplevel('Ticker')

# Build up a new dataframe with AAPL and MSFT returns
return_data = pd.concat([aapl_returns, msft_returns], axis = 1)[1:]
return_data.columns = ['AAPL', 'MSFT']

# add a constant 
X = sm.add_constant(return_data['AAPL'])

# construct the model
model = sm.OLS(return_data['MSFT'], X).fit()

# print the model summary
print(model.summary())
```

```python
# we are using the plot function 
plt.plot(return_data['AAPL'], return_data['MSFT'], 'r.')

# we are using the axis method and linspace method to increase the linespace of the axes
ax = plt.axis()
x = np.linspace(ax[0], ax[1] + 0.01)

# we are using the plot function to specify some methods and values
plt.plot(x, model.params[0] + model.params[1] * x, 'b', lw = 2)

# we are using some methods in the plt function to get our desired graph
plt.grid(True)
plt.axis('tight')
plt.xlabel('Apple Returns')
plt.ylabel('Microsoft returns')

# we are suing the show function to print the graph on the console
plt.show()
```

## Building a Trading Strategy with Python

### There are two types of Trading Strategies generally followed - Momentum and Reversion

#### Momentum Strategy - You do it when you believe that the movement of the stock will continue in its current dirction
#### Reversion Strategy - This strategy departs from the belief that the movement of a quantity will eventually reverse

### The strategy that I am using is called the moving average cross over
### Here I create two simple moving averages of a time series with differing lookback periods i.e. 40 and 100 days
### If the short moving average exceeds the long moving average we go long
### If the long moving average exceeds the short moving average then we exit

```python
# initializing the short and long windows
short_window = 40
long_window = 100

# initializing a new data frame called 'signals' with a column called 'signal'
signals = pd.DataFrame(index=aapl.index)
# we are initializing all values of the signal column to 0.0
signals['signal'] = 0.0

# creating a short simple moving average over the short window 
signals['short_mavg'] = aapl['Close'].rolling(window=short_window, min_periods=1, center=False).mean()

# creating a long simple moving average over the long window
signals['long_mavg'] = aapl['Close'].rolling(window=long_window, min_periods=1, center=False).mean()

# creating signals
signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] 
                                            > signals['long_mavg'][short_window:], 1.0, 0.0)

# generating trading orders
signals['positions'] = signals['signal'].diff()

#printing our dataframe signals
print(signals)
#wow our dataframe looks wonderful in the console
```

```python
# initializing the plot figure
fig = plt.figure()

# adding a sub plot and label for y-axis
ax1 = fig.add_subplot(111, ylabel='Price in $')
# in the above statement 111 indicates, the number of rows

# plotting the closing price
aapl['Close'].plot(ax=ax1, color='r', lw=2.)

# plotting the short and long moving averages
signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)

# plotting the buy signals
ax1.plot(signals.loc[signals.positions == 1.0].index,
        signals.short_mavg[signals.positions == 1.0],
        '^', markersize=10, color='m')

# plotting the sell signals
ax1.plot(signals.loc[signals.positions == -1.0].index,
        signals.short_mavg[signals.positions == -1.0],
        'v', markersize = 10, color = 'k')

# show the plot
plt.show()
```

## Backtesting The Trading Strategy

### A simple backtester consists of a strategy, data handler, a portfolio and a execution handler

### Out of the above four points, we already have a strategy, and we already have a data handler, which is our
### pandas-datareader package that we used to get our saved data in excel into Python

### We will build a portfolio and we will not be looking at an execution handler in this project as it is beyond our
### scope and we are not actually trading in the stock market

### Implementing a simple backtester with pandas

```python
# First let us set the initial capital
initial_capital = float(100000.0)

# creating a dataframe called 'positions'
positions = pd.DataFrame(index=signals.index).fillna(0.0)

# we are buying 100 shares
positions['AAPL'] = 100 * signals['signal']

# initializing the portfolio with value owned
portfolio = positions.multiply(aapl['Adj Close'], axis=0)

# Storing the difference in shares owned
poss_diff = positions.diff()

# Adding 'holdings' to portfolio
portfolio['holdings'] = (positions.multiply(aapl['Adj Close'], axis=0)).sum(axis=1)

# Adding 'cash' to the portfolio
portfolio['cash'] = initial_capital - (poss_diff.multiply(aapl['Adj Close'], axis=0)).sum(axis=1).cumsum()

# Adding 'total' to portfolio
portfolio['total'] = portfolio['cash'] + portfolio['holdings']

# Adding 'returns' to portfolio
portfolio['returns'] = portfolio['total'].pct_change()
```

```python
# using the figure function to create a new figure and storing it in a variable
fig = plt.figure()

ax1 = fig.add_subplot(111, ylabel = 'Portfolio value in $')

# plotting the equity curve in dollars
portfolio['total'].plot(ax=ax1, lw=2.)

# plotting the buy trades against the equity curve
ax1.plot(portfolio.loc[signals.positions == 1.0].index,
        portfolio.total[signals.positions == 1.0],
        '^', markersize = 10, color = 'm')

# plotting the sell trades against the equity curve
ax1.plot(portfolio.loc[signals.positions == -1.0].index,
        portfolio.total[signals.positions == -1.0],
        'v', markersize = 10, color = 'k')

# showing the plot
plt.show()
```
